```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df = pd.read_csv ("movies.csv")
```

```python

df.isna().sum()
```

```python
features = ['genres', 'keywords', 'title' , 'overview','tagline' ]

for feature in features :
    df[feature] = df[feature].fillna('')
```

```python
combined_features = df['genres'] + " "  + df['keywords'] + " "  +df['title'] + " " +  df['overview'] + " "  + df['tagline']  
```

```python
feature_vectorization = TfidfVectorizer()

```

```python
vectorized_features = feature_vectorization.fit_transform(combined_features)
```

```python
print(vectorized_features)
```

```python
similarity = cosine_similarity(vectorized_features)
```

```python
print(similarity)
```

```python
movie_name = input('Enter a movie name you like: ')

```

```python
list_of_all_movies = df['title'].tolist()
list_of_all_movies
```

```python
find_similar_name = difflib.get_close_matches(movie_name, list_of_all_movies)
find_similar_name
```

```python
correct_name = find_similar_name[0]
correct_name
```

```python
index = df[df['title'] == correct_name].index[0]
index
```

```python
similarity_scores = list(enumerate(similarity[index]))
similarity_scores
```

```python
sorted_similarity_scores = sorted(similarity_scores , key = lambda x: x[1] , reverse = True)
```

```python
first_twenty_film = sorted_similarity_scores[1:21]
first_twenty_film
```

```python
print('First 20 movies we suggest: ')
counter = 1

for movie in first_twenty_film:
    movie_index = movie[0]
    
    movie_name = df[df['index'] == movie_index]['title'].values[0]
    print(counter , ". " ,  movie_name)
    counter += 1
```
